## p5

- *Historique* : **Processing** (Java) > **p5\*js** (JavaScript) > **p5py** (Python)
- **p5** est un **environnement de création** multimédia
- Animations en 2D ou 3D, créations sonores ou visuelles, interactions avec l'utilisateur

## setup + draw

```python
from p5 import *
from random import randint

def setup():
    createCanvas(600, 400)     # Dimensions de la fenêtre
    background(250, 237, 205)  # Couleur du fond
    noStroke()                 # Pas de bordure
    frameRate(1)               # Rafraichissement
    
def draw():
    fill(randint(0,255), 0, 0)  # Variations sur le rouge
    rect(randint(0,500), randint(0,300), 100, 100)  # Position (x,y) aléatoire
```

```python
run()
```

```python
stop()
```

### Exercice

`textSize()` permet de choisir la **taille en pixels d'un texte** à afficher, et `text('Eric', 10, 30)` va **écrire** le prénom `Eric` à la position (10,30). Ecrire un programme qui toutes les 10e de seconde affiche votre prénom avec une taille et une position aléatoire à l'écran. Exemple de résultat :
<img width = "500" src="https://blog.univ-angers.fr/mathsinfo/files/2022/06/eric.png">

```python
from p5 import *
from random import randint

def setup():
    createCanvas(600, 400) 
    background(250, 237, 205)
    noStroke()
    frameRate(10)
    
def draw():
    textSize(randint(1,40))
    text('Eric', randint(0,600), randint(0,400))
```

```python
run()
```

```python
stop()
```

### Exercice

`line(x1,y1,x2,y2)` permet de tracer une **ligne** (**trait**) entre les points de coordonnées `(x1,y1)` et `(x2,y2)`. `strokeWeight(n)` pour définir l'**épaisseur** du trait. Ecrire un programme qui toutes les 10e de seconde affiche une ligne d'*épaisseur aléatoire* entre 1 et 10 reliant un *point quelconque du haut de l'écran* à un *point quelconque du bas de l'écran*. Exemple de résultat :
<img width = "500" src="https://blog.univ-angers.fr/mathsinfo/files/2022/06/lignes.png">

```python
from p5 import *
from random import randint

def setup():
    createCanvas(600, 400) 
    background(250, 237, 205)
    frameRate(10)
    
def draw():
    strokeWeight(randint(1,10));
    line(randint(0,600), 0, randint(0,600), 400)
```

```python
run()
```

```python
stop()
```
