## setup + draw

```python
from p5 import *
from random import randint

def setup():
    createCanvas(600, 400)     # Dimensions de la fenêtre
    background(250, 237, 205)  # Couleur du fond
    noStroke()                 # Pas de bordure
    frameRate(1)               # Rafraichissement
    
def draw():
    fill(randint(0,255), 0, 0)  # Variations sur le rouge
    rect(randint(0,500), randint(0,300), 100, 100)  # Position (x,y) aléatoire
```

```python
run()
```

```python
stop()
```

### Exercice

`textSize()` permet de choisir la **taille en pixels d'un texte** à afficher, et `text('Eric', 10, 30)` va **écrire** le prénom `Eric` à la position (10,30). Ecrire un programme qui toutes les 10e de seconde affiche votre prénom avec une taille et une position aléatoire à l'écran. Exemple de résultat :

<img width = "500" src="https://blog.univ-angers.fr/mathsinfo/files/2022/06/eric.png">

```python
from p5 import *
from random import randint

def setup():
    createCanvas(600, 400) 
    background(250, 237, 205)
    noStroke()
    frameRate(10)
    
def draw():
    textSize(randint(1,40))
    text('Eric', randint(0,600), randint(0,400))
```

`line(x1,y1,x2,y2)` permet de tracer une **ligne** (**trait**) entre les points de coordonnées `(x1,y1)` et `(x2,y2)`. `strokeWeight(n)` pour définir l'**épaisseur** du trait. Ecrire un programme qui toutes les 10e de seconde affiche une ligne d'*épaisseur aléatoire* entre 1 et 10 reliant un *point quelconque du haut de l'écran* à un *point quelconque du bas de l'écran*. <img width = "500" src="https://blog.univ-angers.fr/mathsinfo/files/2022/06/lignes.png">

```python
from p5 import *

def setup():
    createCanvas(600, 400) 

def draw():
    background(250, 237, 205)     # Permet d'effacer l'écran à chaque tour
    circle(mouseX, mouseY, 20)    # Cercle de centre la position de la souris et de rayon 20
    if mouseIsPressed: noLoop()   # On arrête de boucler si la souris est enfoncée
    
run()
```

Ainsi que ce second programme où vous devez essayer d'attraper la balle :

```python
from p5 import *
from random import randint

def cercle_alea():
    global x,y      # on veut utiliser x et y dans draw donc variables globales
    x, y = randint(0, width), randint(0, height)    # Centre du cercle
    circle(x, y, 20)

def setup():
    createCanvas(600, 400)
    background(250, 237, 205)
    cercle_alea()                          # on place une balle au hasard
    
def draw():
    if dist(mouseX, mouseY, x, y) < 30:    # vous êtes trop prêt de la balle ? 
        background(250, 237, 205)          # on efface l'écran
        cercle_alea()                      # et on met une nouvelle balle
    if mouseIsPressed: noLoop()            # si clic de la souris on arrête la boucle
    
run()   
```

Créez un programme qui dessine des cercles qui suivent les mouvements de la souris et dont les rayons sont égaux à la distance entre la position de la souris et le centre de la fenêtre de visualisation. Ainsi, lorsque la souris sera ver le milieu les rayons seront petits et inversement ils grandiront lorsque vous irez vers les bords. Voici un visuel possible :
