# Un jeu de logique

## Version 2D

```python
from p5 import *
from random import randint

changer = [[0,1,3,4],[0,1,2],[1,2,4,5],[0,3,6],[1,3,4,5,7],[2,5,8],[3,4,6,7],[6,7,8],[4,5,7,8]]
jeu = []
for i in range(9): jeu.append(randint(0,1))

def posX(i): return 100 + 100 * (i % 3)

def posY(i): return 100 + 100 * (i // 3)

def affiche(jeu):
    for i in range(9):
        coul = (0, 0, 0) if jeu[i] == 1 else (255, 255, 255)
        fill(coul)
        circle(posX(i), posY(i), 80)

def setup():
    global relache
    createCanvas(400, 400) 
    background(250, 237, 205)
    strokeWeight(4)
    frameRate(10)
    affiche(jeu)
    relache = True

def jouer(i):
    global jeu
    for v in changer[i]: jeu[v] = 1 - jeu[v]

def gagner(jeu):
    return jeu == [1,1,1,1,0,1,1,1,1]

def draw():
  global relache    
  if mouseIsPressed:
      if relache:
          x, y = mouseX, mouseY
          relache = False
          for i in range(9):
              if dist(x,y,posX(i), posY(i)) < 40 :
                  jouer(i)
                  affiche(jeu)
                  if gagner(jeu):
                      textAlign(CENTER, CENTER)
                      text("GAGNE", 200, 200)
                      noLoop()
                  break
  else: relache = True     
run()    
```

```python
stop()
```

## Version 3D

```python
from p5 import *
from random import randint

# https://youtu.be/YMiIaRZdMb0

changer = [[0,1,3,4],[0,1,2],[1,2,4,5],[0,3,6],[1,3,4,5,7],[2,5,8],[3,4,6,7],[6,7,8],[4,5,7,8]]
jeu = []

for i in range(9): jeu.append(randint(0,1))

def posX(i): return 100 * (i % 3)

def posY(i): return 100 * (i // 3)

def affiche(jeu, fin = False):
    rotateX(45)
    rotateZ(20)
    translate(-75, -75)
    for i in range(9):
        if jeu[i] == 1:
            if fin:   # Effet si partie gagnée
                coul, hauteur = (0, randint(20, 255) , 0), randint(5, 50)
            else:
                coul, hauteur = (255, 0, 0), 50
        else:
            coul, hauteur = (150, 150, 150), 10
        fill(coul)
        push()
        translate(posX(i), posY(i), hauteur / 2)
        box(60, 60, hauteur)
        pop()

def setup():
    global fin, relache
    createCanvas(600, 600, WEBGL) 
    angleMode(DEGREES)
    frameRate(10)
    strokeWeight(2)
    affiche(jeu)
    relache = True
    fin = False    # fin du jeu ?

def jouer(i):
    global jeu
    for v in changer[i]: jeu[v] = 1 - jeu[v]

def gagner(jeu):
    return jeu == [1,1,1,1,0,1,1,1,1]
 
def draw():
    global fin, relache
    background(0)    
    if keyIsPressed:
        if relache:
           code = keyCode
           relache = False
           if code >= 49 and code <= 57:
             jouer([6,7,8,3,4,5,0,1,2][code - 49])
             if gagner(jeu): fin = True
    else: relache = True
    affiche(jeu, fin) 
    
run()
```
